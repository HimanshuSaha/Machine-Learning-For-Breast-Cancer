```python
#split the data
X=df.iloc[:,1:13].values
Y=df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state =0)
```

```python
#Scaling of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
def model_evaluate(model):
  y_pred = model.predict(X_test)
  print(classification_report(Y_test, y_pred))

  # Confusion matrix
  cf_matrix = confusion_matrix(Y_test, y_pred)
  categories = ['B','M']
  sns.heatmap(cf_matrix, annot = True, fmt = '', xticklabels = categories, yticklabels = categories)
  plt.xlabel("Predicted values")
  plt.ylabel("Actual values")
  plt.title ("Confusion Matrix")
  plt.show()
```

```python
from sklearn import svm
SVC = svm.SVC()
SVC.fit(X_train, Y_train)
model_evaluate(SVC)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0)
LR.fit(X_train, Y_train)
model_evaluate(LR)
y_pred1 = LR.predict(X_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
BNB= BernoulliNB()
BNB.fit(X_train, Y_train)
model_evaluate(BNB)
y_pred2 = BNB.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB= GaussianNB()
GNB.fit(X_train, Y_train)
model_evaluate(GNB)
y_pred3 = GNB.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier(criterion = 'entropy', random_state=0 )
dc.fit(X_train, Y_train)
model_evaluate(dc)
y_pred3 = dc.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(X_train, Y_train)
model_evaluate(neigh)
y_pred3 = neigh.predict(X_test)
```
